## 1 - GET DUMMIES

```python
dummiesMail_category =  pd.get_dummies(train_df.mail_category,prefix='col')
dummiesMail_category.head()
```

### Concate to original DataFrame

```python
train_dfWithDummies = pd.concat([train_df, dummiesMail_category], axis=1)
train_dfWithDummies.head()
```

**
When you need to group by an ID and a categorical Value forming a unique table with no reapeting element <br/>
the column 'Times' represent the number of times the ID and the categorical value shows
**

```python
groupAndDummy = train_df.groupby(["user_id","mail_category"]).size().reset_index(name="Times")\
        .pivot(index='user_id', columns='mail_category', values='Times').reset_index()#.drop(["user_id"],axis=1)
# groupAndDummy.index.names = ["DummiesUniqueById"]
groupAndDummy.columns.name="DummiesUniqueById"
print groupAndDummy.shape
groupAndDummy.head()
```

## 2 - TRANSFORM INTO VALUE


```python
dict_mail_category = dict([ (k,v) for v,k in enumerate(np.unique(train_df['mail_category'].values))])
dict_mail_category
```

```python
train_df_mail_category_with_values = train_df['mail_category'].map(lambda x:dict_mail_category[x])
train_df_mail_category_with_values.head()
```

### Alter the categorical to values into the main Data Frame


```python
train_df['mail_category'] = train_df['mail_category'].map(lambda x:dict_mail_category[x])
train_df.head()
```
